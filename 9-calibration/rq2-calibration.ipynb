```python

rq_1_config = {
    "config_file": "src/main/resources/config_DHZW_rq2-1.toml",
    "parameterset": "src/main/resources/baseline_parameterset/parameterset_rq2-1.csv",
    "parameterset_write_folder" : "../7-simulation-Sim-2APL/",
    "output_path": 'src/main/resources/baseline_parameterset/output-rq2-1.csv',
    "other_args" : "--use_random_seed true"
}
rq1_1_optimiser = BayesianOptimization(
    f=lambda **params: call_vot_simulation(config=rq_1_config, **params),
    pbounds=vot_pbounds,
    random_state=1,
)

rq1_1_optimiser.maximize(
    init_points=VOT_INIT_POINTS,
    n_iter=VOT_N_ITER
)
```
