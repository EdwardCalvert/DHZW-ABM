```python
import re
import numpy as np

text = ""
with open("data/stt_convergence.txt","r") as f:
    text = f.readlines()
text
```

```python
numeric_results = []
for line in text:
    #print(line)
    match = re.search(r"(-?\d+\.\d+)", line)

    if match:
        result = round(float(match.group(1)), 2)
        numeric_results.append(result) 
print(len(numeric_results))
numeric_results
```

```python
import matplotlib.pyplot as plt

training_phase = numeric_results[100:]

y = np.array(training_phase)
x = np.arange(len(y))
plt.plot()

plt.xlabel('Run #')
plt.ylabel('RMSE')
plt.title('ABM convergence')


window = 7
weights = np.ones(window) / window
sma_7 = np.convolve(y, weights, mode='valid')

x_sma = x[window-1:]

plt.plot(x, y, 'o', label='Data', markersize=4, alpha=0.4)
plt.plot(x_sma, sma_7, 'g-', label='7-pt Moving Avg', linewidth=2)
plt.savefig("data/iteration.pdf", format="pdf")  
plt.show()
```
